```python
toKeep = ['SGG_NM','CTPV_NM',
    'BLDG_SRTFRM_NM', 'BLDG_STRCTR_NM', 'BLDG_SRTRF_NM',
    'SMTPR_LCLSF_NM', 'FCLT_PLC_LCLSF_NM', 'FCLT_PLC_SCLSF_NM', 'FCLT_PLC_MCLSF_NM',
    'DSPT_REQ_HR', 'LFDAU_GRNDS_DSTNC', 'HR_UNIT_ARTMP', 'HR_UNIT_HUM'
]
dfList = []
for i in range(5):
    df = pd.read_csv(f'data/mainData/fire/화재 현황_20{19+i}_전국.csv')
    tempDf = df[df['CTPV_NM'] == '서울특별시']
    filterdDf = tempDf[toKeep]
    dfList.append(filterdDf)

dfMerged = pd.concat(dfList, axis = 0)
dfMerged.to_csv('filtered_data\\FireData19_23.csv', encoding='UTF-8')
```
